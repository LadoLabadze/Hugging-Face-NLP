```python
from datetime import datetime, timedelta
from collections import defaultdict
from transformers import pipeline
from alpaca_trade_api import REST



# ✅ Define date range (last 30 days)
end_date = datetime.today()
start_date = end_date - timedelta(days=365)

# ✅ Store results per day
daily_sentiments = {}

# ✅ Loop over each day
for i in range(366):
    day = start_date + timedelta(days=i)
    day_str = day.strftime('%Y-%m-%d')
    next_day_str = (day + timedelta(days=1)).strftime('%Y-%m-%d')

    try:
        # Get news for this day
        news_items = api.get_news(
            symbol="AMZN",
            start=day_str,
            end=next_day_str,
            include_content=True
        )
    except Exception as e:
        print(f"Error fetching news for {day_str}: {e}")
        continue

    sentiments = []

    for item in news_items:
        try:
            text = item._raw.get("headline") or item._raw.get("summary", "")
            if text.strip():
                result = classifier(text)[0]  # e.g., {'label': 'positive', 'score': 0.99}
                sentiments.append(result["label"].upper())
        except Exception as e:
            print(f"Error analyzing one article: {e}")

    # Calculate score
    if sentiments:
        pos = sentiments.count("POSITIVE")
        neg = sentiments.count("NEGATIVE")
        neu = sentiments.count("NEUTRAL")
        total = len(sentiments)
        score = (pos - neg) / total
        daily_sentiments[day_str] = {
            "positive": pos,
            "negative": neg,
            "neutral": neu,
            "total": total,
            "sentiment_score": round(score, 6)
        }

# ✅ Print daily sentiment results
print("\n📈 Daily Sentiment Summary (Last 30 Days):\n")
for date, stats in daily_sentiments.items():
    print(f"{date}: +{stats['positive']} / -{stats['negative']} / ={stats['neutral']} | "
          f"Total: {stats['total']} | Score: {stats['sentiment_score']:.2f}")


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ✅ Prepare DataFrame from daily_sentiments
df = pd.DataFrame.from_dict(daily_sentiments, orient='index')
df.index = pd.to_datetime(df.index)
df = df.sort_index()

# ✅ Plot setup
plt.figure(figsize=(14, 6))
sns.set(style="whitegrid")

# ✅ Fancy line plot
sns.lineplot(data=df, x=df.index, y='sentiment_score', marker='o', linewidth=2.5, color='dodgerblue')

# ✅ Add labels and title
plt.title("📊 Daily Sentiment Score for AMZN Over Last Year", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sentiment Score")
plt.xticks(rotation=45)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# ✅ Show plot
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ✅ Step 1: Convert sentiment dictionary to DataFrame
sentiment_df = pd.DataFrame.from_dict(daily_sentiments, orient='index')
sentiment_df.index = pd.to_datetime(sentiment_df.index)
sentiment_df = sentiment_df.sort_index()

# ✅ Step 2: Get SPY price for same period (last 365 days)
start_date = (pd.to_datetime(sentiment_df.index.min()) - pd.Timedelta(days=1)).strftime('%Y-%m-%d')
end_date = pd.to_datetime(sentiment_df.index.max()).strftime('%Y-%m-%d')

spy = yf.download('SPY', start=start_date, end=end_date)

# 🔧 Ensure single-level columns
if isinstance(spy.columns, pd.MultiIndex):
    spy.columns = spy.columns.get_level_values(0)

# ✅ Keep only 'Close' and rename
spy = spy[['Close']]
spy.rename(columns={"Close": "SPY_Close"}, inplace=True)

# ✅ Step 3: Combine sentiment and SPY data
combined_df = sentiment_df.merge(spy, left_index=True, right_index=True, how='inner')

# ✅ Step 4: Plot
sns.set(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(15, 6))

# 📈 SPY Close (left Y axis)
color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('SPY Close Price', color=color)
ax1.plot(combined_df.index, combined_df['SPY_Close'], color=color, linewidth=2, label='SPY Price')
ax1.tick_params(axis='y', labelcolor=color)

# 📉 Sentiment Score (right Y axis)
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Sentiment Score', color=color)
ax2.plot(combined_df.index, combined_df['sentiment_score'], color=color, linewidth=2, linestyle='--', label='Sentiment Score')
ax2.axhline(0, color='gray', linestyle='--', linewidth=1)
ax2.tick_params(axis='y', labelcolor=color)

# 🧾 Title and layout
plt.title('📊 SPY Price and Daily Sentiment Score (Last 365 Days)')
fig.tight_layout()
plt.show()
```

```python
# ✅ Step 5: Compute daily price change %
combined_df["price_change_pct"] = combined_df["SPY_Close"].pct_change() * 100

# ✅ Step 6: Drop NaNs (from pct_change)
combined_df.dropna(inplace=True)

# ✅ Step 7: Correlation between sentiment and price movement
correlation = combined_df["sentiment_score"].corr(combined_df["price_change_pct"])
print(f"📊 Correlation between Sentiment Score and SPY % Change: {correlation:.4f}")
```

```python
# ✅ Step 1: Daily return of SPY
combined_df["price_pct_change"] = combined_df["SPY_Close"].pct_change()

# ✅ Drop NA
combined_df.dropna(inplace=True)

# ✅ Store correlations
backward_lag_correlations = {}
forward_lag_correlations = {}

# ✅ Include lag = 0
lag_range = range(0, 11)

for lag in lag_range:
    # 🔁 BACKWARD: Past sentiment vs today's return
    if lag == 0:
        sentiment_lag = combined_df["sentiment_score"]
    else:
        sentiment_lag = combined_df["sentiment_score"].shift(lag)

    temp_back = combined_df.copy()
    temp_back[f"sentiment_lag_{lag}d"] = sentiment_lag
    temp_back.dropna(subset=["price_pct_change", f"sentiment_lag_{lag}d"], inplace=True)
    corr_back = temp_back["price_pct_change"].corr(temp_back[f"sentiment_lag_{lag}d"])
    backward_lag_correlations[lag] = round(corr_back, 4)

    # 🔜 FORWARD: Today's sentiment vs future return
    if lag == 0:
        fwd_return = combined_df["price_pct_change"]
    else:
        fwd_return = combined_df["SPY_Close"].pct_change(periods=lag).shift(-lag) * 100

    temp_fwd = combined_df.copy()
    temp_fwd[f"fwd_return_{lag}d"] = fwd_return
    temp_fwd.dropna(subset=["sentiment_score", f"fwd_return_{lag}d"], inplace=True)
    corr_fwd = temp_fwd["sentiment_score"].corr(temp_fwd[f"fwd_return_{lag}d"])
    forward_lag_correlations[lag] = round(corr_fwd, 4)

# ✅ Print results with lag 0 included
print("\n📊 Sentiment vs SPY % Return Correlation:")
print("Lag Day | Past Sentiment → Today Return | Today Sentiment → Future Return")
print("--------|-------------------------------|-------------------------------")
for lag in lag_range:
    print(f"{lag:>7} | {backward_lag_correlations[lag]:>29.4f} | {forward_lag_correlations[lag]:>30.4f}")

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ✅ Combine lag correlation results into single dict
full_lag_correlation = {}

# Add backward lags (-10 to -1)
for lag in range(1, 11):
    full_lag_correlation[-lag] = backward_lag_correlations[lag]

# Add lag 0
full_lag_correlation[0] = backward_lag_correlations[0]

# Add forward lags (+1 to +10)
for lag in range(1, 11):
    full_lag_correlation[lag] = forward_lag_correlations[lag]

# ✅ Convert to DataFrame
corr_df = pd.Series(full_lag_correlation).sort_index()

# ✅ Plot
plt.figure(figsize=(12, 6))
plt.plot(corr_df.index, corr_df.values, marker='o', linestyle='-', color='purple')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='black', linestyle=':', linewidth=1)

# ✅ Labels
plt.title("📊 Correlation: Sentiment Lag vs SPY % Return (-10 to +10 Days)")
plt.xlabel("Lag Days (Negative = Past Sentiment, Positive = Future Return)")
plt.ylabel("Pearson Correlation")
plt.xticks(range(-10, 11))
plt.grid(True)
plt.tight_layout()
plt.show()

```
